```python
from xml.etree.ElementTree import parse
```

```python
document = parse('data/books.xml')
```

```python
dir(document)
```

```python
book = document.find('book')
```

```python
book.attrib['id']
```

```python
[ele for ele in book]
```

```python
[(ele.tag, ele.text) for ele in book]
```

```python
book_dict = dict([(ele.tag, ele.text) for ele in book])
book_dict.update({'id': book.attrib['id']})
book_dict
```

```python
books = []
for book in document.iterfind('book'):
    book_dict = dict([(ele.tag, ele.text) for ele in book])
    book_dict.update({'id': book.attrib['id']})
    books.append(book_dict)
    
with open('books.json','w') as f:
    json.dump(books,f)
```

```python
books = []

for item in document.iterfind('book'):
    book = {
        "id": item.attrib['id'],
        "author": item.findtext('author'),
        "title": item.findtext('title'),
        "genre": item.findtext('genre'),
        "price": item.findtext('price'),
        "publish_date": item.findtext('publish_date'),
        "description": item.findtext('description').replace('\n      ',''),
        
    }
    books.append(book)
    
print(json.dumps(books, indent=2)) 
```
