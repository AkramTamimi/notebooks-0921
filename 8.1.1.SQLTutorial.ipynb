```python
%%sql
select IdProveedor proveedor,
       max(PrecioUnidad) max_price, 
       min(PrecioUnidad) min_price, 
       round(avg(PrecioUnidad)) avg_price
from Productos
where Suspendido = 'False' and PrecioUnidad > 10
group by IdProveedor
order by (IdProveedor)
limit 5
```

```python
%%sql
select IdProveedor proveedor,
       max(PrecioUnidad) max_price, 
       min(PrecioUnidad) min_price, 
       round(avg(PrecioUnidad)) avg_price
from Productos
where Suspendido = 'False' and PrecioUnidad > 10
group by IdProveedor
having avg_price > 50
order by (IdProveedor)
```

## Create view

```python
%%sql
create view PedidosEspaña as
select NombreCompañia, p.*
from Pedidos p inner join Clientes c on c.IdCliente = p.IdCLiente
where Pais = 'España' 
```

```python
%%sql
with PedidosSpain as
(select NombreCompañia, p.*
from Pedidos p inner join Clientes c on c.IdCliente = p.IdCLiente
where Pais = 'España' )
select count(*) from PedidosSpain
```

## Pandas

```python
import pandas as pd
```

```python
db_url='sqlite:///data/tutorial.db'
```

```python
avg_price = 50
query = f'''
select NombreCompañia proveedor,
       max(PrecioUnidad) max_price, 
       min(PrecioUnidad) min_price, 
       round(avg(PrecioUnidad)) avg_price
from Productos p join 
     Proveedores pp on p.IdProveedor = pp.IdProveedor 
where Suspendido = 'False' and PrecioUnidad > 10
group by NombreCompañia
having avg_price > {avg_price}
order by (NombreCompañia)
'''
```

```python
print(query)
```

```python
df = pd.read_sql(query, db_url)
```
